```python
df_list = []
splitted = df_gangseo_gijeog  # 데이터프레임을 직접 사용하거나 조건에 맞게 필터링하세요.

# 새로운 데이터프레임을 만들기 위해 필요한 컬럼을 정의합니다.
column_names = splitted.columns.tolist()

# 예제 데이터: 컬럼 인덱스를 실제 컬럼 이름에 맞게 수정하세요.
new_data_head = {
    '시군구': ['강서구'] * len(splitted),
    '도서명': splitted[column_names[1]].tolist(),
    '저자': splitted[column_names[2]].tolist(),
    '출판사': splitted[column_names[3]].tolist(),
    '발행년도': splitted[column_names[4]].tolist(),
    'ISBN': splitted[column_names[5]].tolist(),
    '부가기호': splitted[column_names[7]].tolist(),
    '주제분류번호': splitted[column_names[9]].tolist(),
    '도서권수': splitted[column_names[10]].tolist(),
    '대출건수': splitted[column_names[11]].tolist(),
    '등록일자': splitted[column_names[12]].tolist()
}

# 새로운 데이터프레임을 생성합니다.
new_df_head = pd.DataFrame(new_data_head)

df_list.append(new_df_head)

# 새로운 데이터프레임을 생성할 때 사용할 딕셔너리의 초기화
new_df = {
    '시군구': [],
    '도서명': [],
    '저자': [],
    '출판사': [],
    '발행년도': [],
    'ISBN': [],
    '부가기호': [],
    '주제분류번호': [],
    '도서권수': [],
    '대출건수': [],
    '등록일자': []
}

# 각 열의 값을 할당합니다.
for column in df_gangseo_gijeog.columns:
    if column in new_df:
        new_df[column] = df_gangseo_gijeog[column].tolist()

# '시군구' 열을 추가합니다.
new_df['시군구'] = ['강서구'] * len(new_df['도서명'])

# 데이터프레임으로 변환합니다.
new_df = pd.DataFrame(new_df)
df_list.append(new_df)

# 데이터프레임을 병합합니다.
df_all = pd.concat(df_list, ignore_index=True)
df_all
```
